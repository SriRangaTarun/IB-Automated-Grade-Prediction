**OBJECTIVE : To give predicted grades to students based on their performances in various tests and quizes throughout               the year. These grades give the student and his/her teacher an idea of where he/she stands in the                     subject and which subjects he/she should improve on.**

**This data contains the marks and predicted grades of many students throughout the year in various subjects. 
  Here, we have the Unit test, Formative test and Mid Term test marks of students** 

**Read training marks and predicted grades of several students**

**The marks are in the form of grade boundaries from 1 to 7 (lower to higher)**

```python
import numpy as np 
import pandas as pd 

import keras
from keras.utils import to_categorical

train_marks_df = pd.read_csv('/Users/srikrishnapaparaju/APR/train_marks.csv')
train_marks = train_marks_df.drop(['predicted_grade'], axis=1)

train_marks = train_marks.values[:, 1:8].reshape((len(train_marks_df), 7, 1)) - 4

max_features = np.max(train_marks, axis=1)
min_features = np.min(train_marks, axis=1)
range_features = max_features - min_features
median_features = np.median(train_marks, axis=1)
mean_features = np.mean(train_marks, axis=1)

train_engineered_features = np.concatenate((max_features, min_features, range_features, median_features, mean_features), axis=1)

train_actual_predicted_grades = (train_marks_df['predicted_grade'] - 1).values
train_actual_predicted_grades = to_categorical(train_actual_predicted_grades)
# Eg. [1, 0, 0, 0, 0, 0, 0] represents a score of 1 and [0, 0, 0, 1, 0, 0, 0] a score of 4
```

**Create a simple LSTM Network model in Keras and train it on the training data**

```python
import tensorflow as tf


from keras.models import Model
from keras.layers import Input, LSTM, Dense, Dropout, concatenate


test_marks = Input((7,1,))
engineered_features_var = Input((5,))


lstm = LSTM(20, recurrent_dropout=0.25, activation='relu')(test_marks)

dense1 = Dense(20, activation='sigmoid')(lstm)
dropout1 = Dropout(0.25)(dense1)

combined_features = concatenate([dropout1, engineered_features_var])
dense2 = Dense(10, activation='sigmoid')(combined_features)
dropout2 = Dropout(0.25)(dense2)

predicted_grades = Dense(7, activation='softmax')(dropout2)


model = Model(inputs=[test_marks, engineered_features_var], outputs=predicted_grades)
model.compile(loss='categorical_crossentropy', optimizer='rmsprop')


model.fit([train_marks, train_engineered_features], train_actual_predicted_grades, epochs=3000) # validation_data=(np.array(val_marks), np.array(val_actual_predicted_grades)))
```

```python
test_marks_df = pd.read_csv('/Users/srikrishnapaparaju/APR/test_marks.csv')
test_marks = test_marks_df.values
test_marks = np.array(test_marks)[:, 3:10].reshape((4, 7, 1)) - 4

max_features = np.max(test_marks, axis=1)
min_features = np.min(test_marks, axis=1)
range_features = max_features - min_features
median_features = np.median(test_marks, axis=1)
mean_features = np.mean(test_marks, axis=1)

test_engineered_features = np.concatenate((max_features, min_features, range_features, median_features, mean_features), axis=1)
```

```python
predictions = model.predict([test_marks, test_engineered_features])
final_predictions = []

for i in range(0, len(test_marks)):
    max_prob_index = list.index(list(predictions[i]), max(predictions[i])) + 1
    final_predictions.append(max_prob_index)    
```
